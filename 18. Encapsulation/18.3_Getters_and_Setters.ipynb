# **18.3 Getters_and_Setters**

Getters and setters control how attributes are accessed and modified. A setter can validate that HP is never negative, and a getter can compute values on the fly. This lesson covers traditional getter/setter methods and when to use them.

```python
class Pokemon:
    def __init__(self, name, hp):
        self._name = name
        self._hp = hp
        self._max_hp = hp
    
    # GETTER
    def get_hp(self):
        return self._hp
    
    # SETTER with validation
    def set_hp(self, value):
        if value < 0:
            self._hp = 0
        elif value > self._max_hp:
            self._hp = self._max_hp
        else:
            self._hp = value

pikachu = Pokemon("Pikachu", 35)

print(f"HP: {pikachu.get_hp()}")

pikachu.set_hp(-10)  # Clamped to 0
print(f"After -10: {pikachu.get_hp()}")

pikachu.set_hp(100)  # Clamped to max
print(f"After 100: {pikachu.get_hp()}")
```

---

## **Summary**

- Getters provide read access
- Setters provide write access with validation
- Naming: `get_attr()` and `set_attr()`
- Verbose but explicit
- Properties are more Pythonic (next lesson)
