# **4.7 Nested Loops**

Nested loops are loops inside other loops - like battling every Pokemon on your team against every Pokemon on an opponent's team! They're perfect for working with 2D data, grids, and complex iterations. Let's master nested loops!

---

## **What are Nested Loops?**

A **nested loop** is a loop inside another loop. The inner loop completes all its iterations for each iteration of the outer loop.

### **Basic Structure:**
```python
for outer_item in outer_sequence:
    for inner_item in inner_sequence:
        # This runs for every combination
```

---

## **Basic Nested Loop**

````python
# Simple example
for i in range(3):
    print(f"Outer loop: {i}")
    for j in range(2):
        print(f"  Inner loop: {j}")
````

````python
# Pokemon example
trainers = ["Ash", "Misty"]
pokemon = ["Pikachu", "Charizard"]

for trainer in trainers:
    print(f"{trainer}'s team:")
    for poke in pokemon:
        print(f"  - {poke}")
````

---

## **How Nested Loops Execute**

````python
# Understanding the execution order
for i in range(1, 4):
    for j in range(1, 4):
        print(f"({i}, {j})", end=" ")
    print()  # New line after inner loop
````

---

## **Nested Loops with Lists**

````python
# Process 2D list (list of lists)
boxes = [
    ["Pikachu", "Charmander", "Squirtle"],
    ["Bulbasaur", "Eevee", "Meowth"],
    ["Jigglypuff", "Psyduck", "Geodude"]
]

for box_num, box in enumerate(boxes, start=1):
    print(f"Box {box_num}:")
    for pokemon in box:
        print(f"  - {pokemon}")
````

## **Common Nested Loop Patterns**

### **Pattern 1: All Combinations**

````python
# Every Pokemon vs every opponent
your_team = ["Pikachu", "Charizard"]
opponent_team = ["Onix", "Blastoise"]

print("All possible matchups:")
for yours in your_team:
    for theirs in opponent_team:
        print(f"  {yours} vs {theirs}")
````

### **Pattern 2: Multiplication Table**

````python
# Create a times table
for i in range(1, 6):
    for j in range(1, 6):
        print(f"{i * j:3}", end=" ")
    print()  # New line
````

### **Pattern 4: Search in 2D**

````python
# Find Pokemon in boxes
boxes = [
    ["Pikachu", "Charmander", "Squirtle"],
    ["Bulbasaur", "Eevee", "Meowth"],
    ["Jigglypuff", "Psyduck", "Geodude"]
]

searching_for = "Eevee"
found = False

for box_num, box in enumerate(boxes, start=1):
    for pos, pokemon in enumerate(box, start=1):
        if pokemon == searching_for:
            print(f"Found {searching_for} in Box {box_num}, Position {pos}")
            found = True
            break
    if found:
        break
````

---

## **Break and Continue in Nested Loops**

````python
# Break only exits innermost loop
for i in range(1, 4):
    print(f"Outer: {i}")
    for j in range(1, 4):
        if j == 2:
            break  # Exits inner loop only
        print(f"  Inner: {j}")
````

````python
# Break outer loop using flag
found = False
for i in range(1, 4):
    for j in range(1, 4):
        if i == 2 and j == 2:
            found = True
            break  # Exit inner loop
    if found:
        break  # Exit outer loop
````

## **Practical Examples**

### **Example 1: Team vs Team Battle**

````python
# Simulate team battles
your_team = ["Pikachu", "Charizard", "Blastoise"]
opponent_team = ["Onix", "Venusaur", "Arcanine"]

print("=" * 40)
print("        TEAM BATTLE RESULTS")
print("=" * 40)

your_wins = 0
opponent_wins = 0

for i, yours in enumerate(your_team, start=1):
    theirs = opponent_team[i - 1]
    print(f"\nBattle {i}: {yours} vs {theirs}")
    
    # Simulate battle (for demo)
    if i % 2 == 1:
        print(f"  Winner: {yours}")
        your_wins += 1
    else:
        print(f"  Winner: {theirs}")
        opponent_wins += 1

print("\n" + "=" * 40)
print(f"Final Score: You {your_wins} - {opponent_wins} Opponent")
print("=" * 40)
````

### **Example 2: Type Effectiveness Chart**

````python
# Check type effectiveness
attacking_types = ["Fire", "Water", "Grass"]
defending_types = ["Grass", "Fire", "Water"]

print("Type Effectiveness Chart:")
print("=" * 40)

for attack in attacking_types:
    print(f"{attack} attacks:")
    for defend in defending_types:
        # Simple effectiveness logic
        if (attack == "Fire" and defend == "Grass") or \
           (attack == "Water" and defend == "Fire") or \
           (attack == "Grass" and defend == "Water"):
            effect = "Super Effective!"
        elif attack == defend:
            effect = "Not very effective"
        else:
            effect = "Normal"
        
        print(f"  vs {defend}: {effect}")
````
