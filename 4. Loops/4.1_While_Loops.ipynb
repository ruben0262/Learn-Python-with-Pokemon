# **4.1 While Loops**

While loops let you repeat code as long as a condition is True. Like a Pokemon trainer battling until they win or run out of Pokemon, while loops continue until a condition becomes False. Let's master repetition with while loops!

---

## **What is a While Loop?**

A **while loop** repeats a block of code as long as a condition remains True.

### **Syntax:**

```python
while condition:
    # code to repeat
    # update condition eventually!
```

---

## **Basic While Loop**

````python
# Count from 1 to 5
count = 1

while count <= 5:
    print(f"Count: {count}")
    count += 1  # IMPORTANT: Update the counter!

print("Done!")
````

### **How It Works:**

1. Check if condition is True
2. If True, run the code block
3. Go back to step 1
4. If False, exit the loop

````python
# Pokemon leveling up
level = 1

while level < 5:
    print(f"Pikachu is level {level}")
    level += 1

print(f"Pikachu reached level {level}!")
````

---

## **Infinite Loops (Be Careful!)**

If the condition never becomes False, the loop runs forever!

````python
# DON'T RUN THIS - It's an infinite loop!
# count = 1
# while count <= 5:
#     print(count)
#     # Forgot to increment! Loop never ends!

# CORRECT VERSION:
count = 1
while count <= 5:
    print(count)
    count += 1  # This makes the loop eventually stop
````

### **Golden Rule:** Always make sure your loop condition will eventually become False!

---

## **Common While Loop Patterns**

### **Pattern 1: Count Up**

````python
# Count from 1 to 10
num = 1
while num <= 10:
    print(num)
    num += 1
````

### **Pattern 2: Count Down**

````python
# Countdown from 5 to 1
countdown = 5
while countdown > 0:
    print(f"{countdown}...")
    countdown -= 1
print("Blast off! ðŸš€")
````

### **Pattern 3: Until Condition Met**

````python
# Gain experience until level up
exp = 0
exp_needed = 100

while exp < exp_needed:
    exp += 25  # Gain 25 exp per battle
    print(f"Gained experience! Total: {exp}")

print("Level up!")
````

### **Pattern 4: User Input Loop**

````python
# Keep asking until valid input (demo only)
# In practice, you'd use actual input()

choice = "invalid"
valid_choices = ["attack", "defend", "item"]

# Simulating user eventually choosing "attack"
attempts = 0
while choice not in valid_choices:
    attempts += 1
    if attempts == 1:
        choice = "jump"  # Invalid
    elif attempts == 2:
        choice = "run"   # Invalid
    else:
        choice = "attack"  # Valid!
    
    if choice not in valid_choices:
        print(f"'{choice}' is not valid. Try again.")

print(f"You chose: {choice}")
````

## **While Loop with Else**

A while loop can have an `else` clause that runs when the loop completes normally (not via break):

````python
# Search for shiny Pokemon
encounters = 0
found_shiny = False

while encounters < 5:
    encounters += 1
    print(f"Encounter #{encounters}")
    # Simulating: shiny not found
else:
    print("Searched all encounters, no shiny found")
````

## **Practical Examples**

### **Example 1: Pokemon Battle**

````python
# Battle until one Pokemon faints
pikachu_hp = 50
opponent_hp = 45
turn = 0

while pikachu_hp > 0 and opponent_hp > 0:
    turn += 1
    print(f"\n--- Turn {turn} ---")
    
    # Pikachu attacks
    damage = 15
    opponent_hp -= damage
    opponent_hp = max(0, opponent_hp)  # Don't go below 0
    print(f"Pikachu deals {damage} damage!")
    print(f"Opponent HP: {opponent_hp}")
    
    if opponent_hp > 0:
        # Opponent attacks back
        damage = 12
        pikachu_hp -= damage
        pikachu_hp = max(0, pikachu_hp)
        print(f"Opponent deals {damage} damage!")
        print(f"Pikachu HP: {pikachu_hp}")

# Battle result
if pikachu_hp > 0:
    print("\nðŸŽ‰ Pikachu wins!")
else:
    print("\nðŸ’€ Pikachu fainted!")
````

### **Example 2: Leveling System**

````python
# Gain levels through experience
level = 1
exp = 0

while level < 5:
    exp_needed = level * 100  # More exp needed per level
    print(f"\nLevel {level}: Need {exp_needed} total EXP")
    
    while exp < exp_needed:
        exp += 35  # Gain exp per battle
        print(f"  Gained EXP! Current: {exp}/{exp_needed}")
    
    level += 1
    print(f"  â­ Leveled up to {level}!")

print(f"\nReached level {level}!")
````

### **Example 3: Catching Pokemon**

````python
# Try to catch a Pokemon
pokemon_caught = False
pokeballs = 5
catch_chance = 0.3  # 30% chance per throw

attempt = 0
while not pokemon_caught and pokeballs > 0:
    attempt += 1
    pokeballs -= 1
    
    print(f"\nAttempt {attempt}: Throwing Pokeball...")
    
    # Simulate catch (for demo, let's say it takes 3 attempts)
    if attempt >= 3:
        pokemon_caught = True
        print("Gotcha! Charizard was caught!")
    else:
        print("Oh no! The Pokemon broke free!")
        print(f"Pokeballs remaining: {pokeballs}")

if not pokemon_caught:
    print("\nRan out of Pokeballs!")
````

## **While Loop vs For Loop**

**Use While When:**
- You don't know how many iterations needed
- Loop depends on a changing condition
- Examples: user input validation, game loops, waiting for events

**Use For When:**
- You know the number of iterations
- Iterating through a collection
- Examples: processing a list, repeating N times

We'll learn for loops in the next lesson!
