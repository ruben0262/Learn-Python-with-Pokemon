# **17.3 Super_Function**

The `super()` function calls methods from the parent class. When overriding `__init__`, use `super().__init__()` to run the parent's initialization, then add child-specific setup. This eliminates code duplication and ensures proper initialization chains.

```python
class Pokemon:
    def __init__(self, name, level):
        self.name = name
        self.level = level

class ElectricPokemon(Pokemon):
    def __init__(self, name, level, voltage):
        super().__init__(name, level)  # Call parent's __init__
        self.voltage = voltage

pikachu = ElectricPokemon("Pikachu", 25, 100)
print(f"{pikachu.name} Lv.{pikachu.level} ({pikachu.voltage}V)")
```

---

## **Summary**

- `super()` accesses parent class
- `super().__init__()` calls parent's constructor
- `super().method()` calls parent's method
- Avoids code duplication
- Maintains initialization chains
