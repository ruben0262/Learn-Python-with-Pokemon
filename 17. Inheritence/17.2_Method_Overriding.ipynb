# **17.2 Method_Overriding**

Method overriding lets child classes replace parent methods with their own implementations. When Pikachu attacks, it uses Thunderbolt instead of a basic attack. The child's version runs, not the parent's. In this lesson you'll learn to override methods, when to override, and how to completely replace parent behavior while maintaining the same interface.

---

## **Basic Method Overriding**

```python
class Pokemon:
    def __init__(self, name):
        self.name = name
    
    def attack(self):
        return f"{self.name} uses basic attack!"

class ElectricPokemon(Pokemon):
    def attack(self):  # Override parent's attack
        return f"{self.name} uses Thunderbolt! âš¡"

# Parent version
generic = Pokemon("MissingNo")
print(generic.attack())

# Child's overridden version
pikachu = ElectricPokemon("Pikachu")
print(pikachu.attack())
```

---

## **Summary**

- Override methods by defining them in child class
- Child's version replaces parent's
- Maintains same method signature
- Enables polymorphism
