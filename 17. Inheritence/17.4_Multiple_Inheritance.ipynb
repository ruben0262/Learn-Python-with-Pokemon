# **17.4 Multiple_Inheritance**

Python allows a class to inherit from multiple parents. A `FlyingFirePokemon` can inherit from both `FlyingPokemon` and `FirePokemon`, gaining both fly() and burn() abilities. This lesson covers multiple inheritance syntax, the diamond problem, and when to use (or avoid) it.

```python
class FlyingPokemon:
    def fly(self):
        return f"{self.name} flies!"

class FirePokemon:
    def burn(self):
        return f"{self.name} uses Flamethrower!"

class Charizard(FlyingPokemon, FirePokemon):  # Multiple parents
    def __init__(self, name):
        self.name = name

charizard = Charizard("Charizard")
print(charizard.fly())   # From FlyingPokemon
print(charizard.burn())  # From FirePokemon
```

---

## **Summary**

- Syntax: `class Child(Parent1, Parent2):`
- Inherits from all parents
- Order matters (MRO)
- Use sparingly â€” often a design smell
- Consider composition instead
